```python
# Checking VIF Values for multicolinearity.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

```python
# Dropping the column has high vif values.
x.drop('DATE',axis=1,inplace=True)
```

```python
# Checking the VIF Values again
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

Now Multicolinearity reduced VIF values are acceptable range. So Lets proceed to build a model.

## Machine Learning Models

```python
# Importing necessary libraries for model training and testing.
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,AdaBoostRegressor,GradientBoostingRegressor,BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import Lasso,Ridge
```

```python
# Fing the best random state for better r2 score.
maxaccu=0
maxrs=0
for i in range(1,246):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred_lr=lr.predict(x_test)
    acc=r2_score(y_test,pred_lr)
    if acc>maxaccu:
        maxaccu=acc
        maxrs=i
print("Maximum r2 score : ",maxaccu," on random state : ",maxrs)
```

Random State 169 gives the highest r2 score out of 246 random state. So lets split the dataset training and testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=maxrs,test_size=.20)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
pred_LR=LR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_LR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_LR))
print("Mean squared error ",mean_squared_error(y_test,pred_LR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_LR)))
print("\n")
cvs=cross_val_score(LR,x,y).mean()
print("Cross Val score : ",cvs)
print("The Difference between R2 score and cross val score ",r2_score(y_test,pred_LR)-cvs)
```
