```python
# Seperating features and target.
x=df1.drop('W',axis=1)
y=df1['W']
```

```python
# Doing standard scaling 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
# Checking VIF Values of each column.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

By seeing this VIF value of 'ER' column gives the more VIF values so lets drop the column & will check the VIF value again.

```python
x=x.drop('ER',axis=1)
```

```python
x
```

```python
vif=pd.DataFrame()
vif['Feature']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

Now VIF value is reduced lets proceed with same VIF Value if we drop one more column again it leads to data loss.

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,AdaBoostRegressor,GradientBoostingRegressor,BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import Lasso,Ridge
```

```python
maxaccu=0
maxrs=0
for i in range(1,29):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred_lr=lr.predict(x_test)
    acc=r2_score(y_test,pred_lr)
    if acc>maxaccu:
        maxaccu=acc
        maxrs=i
print("Maximum r2 score : ",maxaccu," on random state : ",maxrs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=maxrs,test_size=.20)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
pred_LR=LR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_LR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_LR))
print("Mean squared error ",mean_squared_error(y_test,pred_LR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_LR)))
```

```python
GBR=GradientBoostingRegressor()
GBR.fit(x_train,y_train)
pred_GBR=GBR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_GBR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_GBR))
print("Mean squared error ",mean_squared_error(y_test,pred_GBR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_GBR)))
```

```python
lso=Lasso()
lso.fit(x_train,y_train)
pred_lso=lso.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_lso))
print("Mean absolute error ",mean_absolute_error(y_test,pred_lso))
print("Mean squared error ",mean_squared_error(y_test,pred_lso))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_lso)))
```

```python
rdg=Ridge()
rdg.fit(x_train,y_train)
pred_rdg=rdg.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_rdg))
print("Mean absolute error ",mean_absolute_error(y_test,pred_rdg))
print("Mean squared error ",mean_squared_error(y_test,pred_rdg))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_rdg)))
```

```python
knr=KNeighborsRegressor()
knr.fit(x_train,y_train)
pred_knr=knr.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_knr))
print("Mean absolute error ",mean_absolute_error(y_test,pred_knr))
print("Mean squared error ",mean_squared_error(y_test,pred_knr))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_knr)))
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
pred_RFR=RFR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_RFR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_RFR))
print("Mean squared error ",mean_squared_error(y_test,pred_RFR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_RFR)))
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
pred_DTR=DTR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_DTR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_DTR))
print("Mean squared error ",mean_squared_error(y_test,pred_DTR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_DTR)))
```

```python
ETR=ExtraTreesRegressor()
ETR.fit(x_train,y_train)
pred_ETR=ETR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_ETR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_ETR))
print("Mean squared error ",mean_squared_error(y_test,pred_ETR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_ETR)))
```

```python
BR=BaggingRegressor()
BR.fit(x_train,y_train)
pred_BR=BR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_BR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_BR))
print("Mean squared error ",mean_squared_error(y_test,pred_BR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_BR)))
```

```python
ABR=AdaBoostRegressor()
ABR.fit(x_train,y_train)
pred_ABR=ABR.predict(x_test)
print("R2 Score ",r2_score(y_test,pred_ABR))
print("Mean absolute error ",mean_absolute_error(y_test,pred_ABR))
print("Mean squared error ",mean_squared_error(y_test,pred_ABR))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,pred_ABR)))
```

Out of all models linear Regression model is working wel so lets check the cross validation score and hyper parameter tuning to improve the model perfomance 

```python
"Linear Regression is giving 0.94, rdg is giving 0.83 and Gradient boosting regressor is giving 0.83 so lets check the cross val score of each model"

score=cross_val_score(LR,x,y)
print(score)
print(score.mean())
print("Difference between r2 score and cross val score : ",r2_score(y_test,pred_LR)-score.mean())
```

```python
score=cross_val_score(rdg,x,y)
print(score)
print(score.mean())
print("Difference between r2 score and cross val score : ",r2_score(y_test,pred_rdg)-score.mean())
```

```python
score=cross_val_score(GBR,x,y)
print(score)
print(score.mean())
print("Difference between r2 score and cross val score : ",r2_score(y_test,pred_GBR)-score.mean())
```

By seeing this cross validation of three models ridge model is giving less cross val score but Linear regression is giving high accuracy and cross val score also less than 1. Its in acceptable range.
So we will consider Linear Regression is best performing & final model. Lets do the hyper parameter tuning by using GridsearchCV ti improve the model performance.

```python
# Hyperparameter tuning using GridsearchCV
from sklearn.model_selection import GridSearchCV       
parameter={'fit_intercept':[True,bool],
           'n_jobs':[-1,1,2]
           }
```

```python
gscv=GridSearchCV(LinearRegression(),parameter,cv=5)
gscv.fit(x_train,y_train)
```

```python
gscv.best_params_
```

```python
final_model=LinearRegression(fit_intercept=True, n_jobs=-1)        # Hyper parameter tuning.
final_model.fit(x_train,y_train)
final_predict=final_model.predict(x_test)
acc=r2_score(y_test,final_predict)                         # Linear Regression performing wel.
print(acc*100)
```

After applying Hyper parameter tuning model performance is not improving much so lets consider this as final model and will save the model.

```python
import joblib
joblib.dump(final_model,"baseball.pkl")           # Saving the model

model=joblib.load("baseball.pkl")
prediction=model.predict(x_test)                 # Reloading back the model
prediction

a=np.array(y_test)                                # Predicting the saved model
data1=pd.DataFrame()
data1['Predicted']=prediction
data1['Original']=a
data1
```
