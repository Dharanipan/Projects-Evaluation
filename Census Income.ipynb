```python
# Seperating features and target or label as x and y.
x=df1.drop('Income',axis=1)
y=df1['Income']
```

```python
x.shape
```

```python
y.shape
```

```python
# Doing standard scaling to avoid the biasness.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x.values),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Feature']=x.columns
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif
```

Multicolinearity is present but its very less vif values and these values are acceptable range. So lets proceed with these values to train the model.

```python
y.value_counts()  # Data imbalace is there so applying over sampling technique using smote method.
```

```python
# Doing over sampling by using smote method.
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Finding the best accuracy score in random state
maxacc=0
maxrs=0
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    predict_lr=lr.predict(x_test)
    acc=accuracy_score(y_test,predict_lr)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print("Best accuracy score is ",maxacc,'at random state',maxrs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=maxrs,test_size=0.20)
```

```python
from sklearn.model_selection import cross_val_score
LR=LogisticRegression()
LR.fit(x_train,y_train)
predict_LR=LR.predict(x_test)
print("Accuracy score ",accuracy_score(y_test,predict_LR))
print("Confusion matrix ",confusion_matrix(y_test,predict_LR))
print("Classification Report ",classification_report(y_test,predict_LR))
print("\n")
cvs=cross_val_score(LR,x,y,cv=5)
print("Cross val score ",cvs)
print("Cross val score mean ",cvs.mean())
print("Difference between accuracy score and cross val score : ",accuracy_score(y_test,predict_LR)-cvs.mean())
```

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predict_RFC=RFC.predict(x_test)
print("Accuracy score ",accuracy_score(y_test,predict_RFC))
print("Confusion matrix ",confusion_matrix(y_test,predict_RFC))
print("Classification Report ",classification_report(y_test,predict_RFC))
print("\n")
cvs=cross_val_score(RFC,x,y,cv=5)
print("Cross val score ",cvs)
print("Cross val score mean ",cvs.mean())
print("Difference between accuracy score and cross val score : ",accuracy_score(y_test,predict_RFC)-cvs.mean())
```

```python
ETC=ExtraTreesClassifier()
ETC.fit(x_train,y_train)
predict_ETC=ETC.predict(x_test)
print("Accuracy score ",accuracy_score(y_test,predict_ETC))
print("Confusion matrix ",confusion_matrix(y_test,predict_ETC))
print("Classification Report ",classification_report(y_test,predict_ETC))
print("\n")
cvs=cross_val_score(ETC,x,y,cv=5)
print("Cross val score ",cvs)
print("Cross val score mean ",cvs.mean())
print("Difference between accuracy score and cross val score : ",accuracy_score(y_test,predict_ETC)-cvs.mean())
```

```python
GBC=GradientBoostingClassifier()
GBC.fit(x_train,y_train)
predict_GBC=GBC.predict(x_test)
print("Accuracy score ",accuracy_score(y_test,predict_GBC))
print("Confusion matrix ",confusion_matrix(y_test,predict_GBC))
print("Classification Report ",classification_report(y_test,predict_GBC))
print("\n")
cvs=cross_val_score(GBC,x,y,cv=5)
print("Cross val score ",cvs)
print("Cross val score mean ",cvs.mean())
print("Difference between accuracy score and cross val score : ",accuracy_score(y_test,predict_GBC)-cvs.mean())
```

```python
ABC=AdaBoostClassifier()
ABC.fit(x_train,y_train)
predict_ABC=ABC.predict(x_test)
print("Accuracy score ",accuracy_score(y_test,predict_ABC))
print("Confusion matrix ",confusion_matrix(y_test,predict_ABC))
print("Classification Report ",classification_report(y_test,predict_ABC))
print("\n")
cvs=cross_val_score(ABC,x,y,cv=5)
print("Cross val score ",cvs)
print("Cross val score mean ",cvs.mean())
print("Difference between accuracy score and cross val score : ",accuracy_score(y_test,predict_ABC)-cvs.mean())
```

Out of all models Extra Trees Classifier is giving high accuracy. so lets consider this as best model and will improve the model performance by using hyperparameter tuning.

```python
# Using hyperparameter tuning.
from sklearn.model_selection import GridSearchCV
parameters={'criterion':['gini','entropy','log_loss']}
gscv=GridSearchCV(ExtraTreesClassifier(),parameters,cv=5)
gscv.fit(x_train,y_train)
gscv.best_params_
```

```python
final_model=ExtraTreesClassifier(criterion='entropy')
final_model.fit(x_train,y_train)
predict_fin_modl=final_model.predict(x_test)
acc=accuracy_score(y_test,predict_fin_modl)
print(acc*100)
```

```python
y_pred_proba=final_model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_proba)
plt.plot([0,1],[0,1],'k--')                               # Plotting AUC ROC Curve
plt.plot(fpr,tpr,label='Extra Trees Classifier') 
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
auc_score=roc_auc_score(y_test,final_model.predict(x_test))
print("Auc score : ",auc_score)
```
